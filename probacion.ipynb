```python
import joblib
import numpy as np

# Cargamos el modelo entrenado
modelo = joblib.load("modelo_logistic_regression.pkl")

print("ğŸ¤– Â¡Bienvenido al Predictor de Riesgo de DeserciÃ³n!")

# Solicitamos datos al usuario
prom_gral = float(input("ğŸ“˜ Ingresa el promedio general del alumno: "))
asistencia = float(input("ğŸ“… Ingresa el porcentaje de asistencia: "))
edad = int(input("ğŸ‚ Ingresa la edad del alumno: "))
dependencia = int(input("ğŸ« Ingresa el cÃ³digo de dependencia (1=Muni, 2=Particular, 3=Subvencionado): "))

# Convertimos los datos a un array 2D para el modelo
nuevos_datos = np.array([[prom_gral, asistencia, edad, dependencia]])

# Realizamos la predicciÃ³n
prediccion = modelo.predict(nuevos_datos)[0]
probabilidad = modelo.predict_proba(nuevos_datos)[0][1]

# Mostramos resultado
if prediccion == 1:
    print(f"âš ï¸ Riesgo de deserciÃ³n detectado ({probabilidad:.2%} de probabilidad)")
else:
    print(f"âœ… Sin riesgo de deserciÃ³n ({probabilidad:.2%} de probabilidad)")
```
